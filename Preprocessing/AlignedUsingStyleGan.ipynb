```python
import os
import shutil

def copy_files_with_extras(source_dir, destination_dir, extras_dir):
    """Copies files from source_dir to destination_dir.
    If a file already exists in destination_dir, it's copied to extras_dir.

    Args:
        source_dir: The directory to copy files from.
        destination_dir: The directory to copy files to.
        extras_dir: The directory to store existing files.
    """
    # Create destination and extras directories if they don't exist
    os.makedirs(destination_dir, exist_ok=True)
    os.makedirs(extras_dir, exist_ok=True)

    # Iterate through all files and subdirectories in the source directory
    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)
        destination_path = os.path.join(destination_dir, filename)
        extras_path = os.path.join(extras_dir, filename)

        # If the file doesn't exist in the destination, copy it
        if not os.path.exists(destination_path):
            shutil.copy2(source_path, destination_path)
        # If the file already exists, copy it to the extras directory
        else:
            shutil.copy2(source_path, extras_path)

# Call the function with your directories
source_dir = "/content/drive/MyDrive/Child Generator/Testing/Testing_leftovers"
destination_dir = "/content/drive/MyDrive/Child Generator/Testing"
extras_dir = "/content/drive/MyDrive/Child Generator/extras"
copy_files_with_extras(source_dir, destination_dir, extras_dir)
```

```python
import os
import shutil

def copy_files_back(source_dir, destination_dir):
    """Copies files from source_dir to destination_dir.

    Args:
        source_dir: The directory to copy files from.
        destination_dir: The directory to copy files to.
    """
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_dir, exist_ok=True)

    # Iterate through all files and subdirectories in the source directory
    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)
        destination_path = os.path.join(destination_dir, filename)

        # Copy the file to the destination directory
        if os.path.isfile(source_path):
            shutil.copy2(source_path, destination_path)
        # Copy the subdirectory to the destination directory (recursively)
        elif os.path.isdir(source_path):
            shutil.copytree(source_path, destination_path)

# Call the function with the original source and destination directories
source_dir = "/content/Real-ESRGAN/input_images"
destination_dir = "/content/drive/MyDrive/Child Generator/Families-edited-photosonly"
copy_files_back(source_dir, destination_dir)
```

```python
# prompt: count how many files are in this folder /content/drive/MyDrive/Child Generator/Testing

import os

def count_files(directory):
  """Counts the number of files in a given directory.

  Args:
    directory: The path to the directory.

  Returns:
    The number of files in the directory.
  """
  count = 0
  for _, _, files in os.walk(directory):
    count += len(files)
  return count

# Count the number of files in the specified directory
directory_path = "/content/drive/MyDrive/Child Generator/Testing"
file_count = count_files(directory_path)

print(f"Number of files in '{directory_path}': {file_count}")
```

```python
# prompt: I want to recreate the checkpoint3.csv for each row, I want to query the father_name in my member_image_df['member'] == father_name and repopulate the father_images from the checkpoint csv

# Assuming 'df' is your checkpoint3.csv DataFrame and 'member_image_df' is your member_image.csv DataFrame

# Iterate through the rows of the checkpoint3 DataFrame
path_to_csv = '/content/drive/MyDrive/Child Generator/CSVs/checkpoint3.csv'
checkpoint_df = pd.read_csv(path_to_csv)
print(checkpoint_df.head())

path_to_csv = '/content/drive/MyDrive/Child Generator/CSVs/member_image2.csv'
member_image_df = pd.read_csv(path_to_csv)
print(member_image_df.head())

for index, row in checkpoint_df.iterrows():
    father_name = row['father_name']
    mother_name = row['mother_name']
    child_name = row['child_name']

    # Create a list to store the image paths for the father
    father_images = []
    mother_images = []
    child_images = []

    # Query the member_image_df for the father's image paths
    father_df = member_image_df[member_image_df['member'] == father_name][['image path']]
    mother_df = member_image_df[member_image_df['member'] == mother_name][['image path']]
    child_df = member_image_df[member_image_df['member'] == child_name][['image path']]

    # Check if the DataFrames are empty before accessing elements
    father_images = father_df.iloc[0, 0] if not father_df.empty else []
    mother_images = mother_df.iloc[0, 0] if not mother_df.empty else []
    child_images = child_df.iloc[0, 0] if not child_df.empty else []

    father_images_eval = eval(father_images) if isinstance(father_images, str) else father_images
    mother_images_eval = eval(mother_images) if isinstance(mother_images, str) else mother_images
    child_images_eval = eval(child_images) if isinstance(child_images, str) else child_images

    # Update the 'father_images' column in the current row of the checkpoint3 DataFrame
    checkpoint_df.at[index, 'father_images'] = father_images_eval
    checkpoint_df.at[index, 'mother_images'] = mother_images_eval
    checkpoint_df.at[index, 'child_images'] = child_images_eval




print(checkpoint_df.head())
# Save the updated DataFrame to a new CSV file
checkpoint_df.to_csv('/content/drive/MyDrive/Child Generator/CSVs/checkpoint10.csv', index=False)
```
